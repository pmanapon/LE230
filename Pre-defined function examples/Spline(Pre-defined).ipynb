# Example of Spline interpolation using pre-defined function

```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
def f(x):
    return np.sin(x) + np.sin(1.2 * x)
```

```python
xdata = np.array([1,2,3,4,5,7,9,12,14,15,16,17])
ydata = f(xdata)

xnew = np.arange(xdata[0],xdata[-1],0.001)
yfnc = f(xnew)
warnings.filterwarnings('ignore')
y_order1 = interpolate.spline(xdata, ydata, xnew, 1)
y_order2 = interpolate.spline(xdata, ydata, xnew, 2)
y_order3 = interpolate.spline(xdata, ydata, xnew, 3)
```

```python
plt.clf()
plt.plot(xdata, ydata,'ko', label= 'Data points')
plt.plot(xnew, yfnc,'red', label= 'f(x)')
plt.plot(xnew, y_order1,'green', label= 'Linear')
plt.plot(xnew, y_order2,'blue', label= 'Quadratic')
plt.plot(xnew, y_order3,'orange', label= 'Cubic')
plt.legend(loc='lower right')
plt.title("Spline Interpolation")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
