# Example of Lagrange interpolation using pre-defined function

```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x) + np.sin(1.2 * x)
```

```python
xdata = np.array([1,2,3,4,5,7,9,12,14,15,16,17])
ydata = f(xdata)

l = interpolate.lagrange(xdata, ydata)
print(l)

xnew = np.arange(xdata[0],xdata[-1],0.001)
ynew = l(xnew)
yfnc = f(xnew)
```

```python
plt.clf()
plt.plot(xdata, ydata,'ko', label= 'Data points')
plt.plot(xnew, yfnc,'red', label= 'f(x)')
plt.plot(xnew, ynew,'blue', label= 'Interpolation polynomial')
plt.legend(loc='lower right')
plt.title("Lagrange Interpolation")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
